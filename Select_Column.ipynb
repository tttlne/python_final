```python
import pandas as pd
import numpy as np
```

```python
# Đọc dữ liệu
df=pd.read_excel("/content/drive/MyDrive/Colab_Notebooks/napbook.vn.xlsx")
df.info()
```

```python
# Xóa các cột không cần thiết
df = df.drop(columns=['text', 'shared_text', 'original_text', 'image_lowquality', 'images','images_description', 'images_lowquality', 'images_lowquality_description',
                     'video_duration_seconds', 'video_height', 'video_id', 'video_quality', 'video_size_MB', 'video_thumbnail', 'video_watches',
                     'video_width', 'likes', 'link', 'links', 'user_id', 'username', 'user_url', 'is_live', 'factcheck', 'shared_post_id', 'shared_time',
                     'shared_user_id', 'shared_username', 'shared_post_url', 'available', 'reactors', 'w3_fb_url', 'with', 'page_id', 'sharers', 'image_id',
                     'image_ids', 'was_live', 'fetched_time'])
```

```python
df.info()
```

```python
for i in range(0, len(df['video'].values), 1):
    if pd.isna(df['video'].values[i]):
        df['video'].values[i] = 0
    if pd.isna(df['image'].values[i]):
        df['image'].values[i] = 0
```

```python
# Tạo cột giờ đăng bài
df['hour_time'] = df['time'].dt.hour
df
```

```python
# Tạo file
df.to_excel("/content/drive/MyDrive/Colab_Notebooks/napbook.vn_dropped.xlsx")
```

```python
# Tách các loại reactions
df['reactions'] = df['reactions'].apply(lambda x : dict(eval(x)) )
df_reactions = df['reactions'].apply(pd.Series )
df_reactions
```

```python
# Thay giá trị NaN bằng 0
for i in range(0, len(df_reactions['thích'].values), 1):
    if np.isnan(df_reactions['thích'].values[i]):
        df_reactions['thích'].values[i] = 0

    if np.isnan(df_reactions['yêu thích'].values[i]):
        df_reactions['yêu thích'].values[i] = 0

    if np.isnan(df_reactions['haha'].values[i]):
        df_reactions['haha'].values[i] = 0

    if np.isnan(df_reactions['wow'].values[i]):
        df_reactions['wow'].values[i] = 0

    if np.isnan(df_reactions['buồn'].values[i]):
        df_reactions['buồn'].values[i] = 0

    if np.isnan(df_reactions['phẫn nộ'].values[i]):
        df_reactions['phẫn nộ'].values[i] = 0

    if np.isnan(df_reactions['thương thương'].values[i]):
        df_reactions['thương thương'].values[i] = 0

df_reactions
```

```python
# Gộp các cột lại vào dataframe
df_reactions = pd.concat([df, df_reactions], axis=1).drop('reactions', axis=1)
df_reactions
```
