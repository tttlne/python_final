## B√†i post c√≥ l∆∞·ª£t t∆∞∆°ng t√°c l·ªõn nh·∫•t

```python
# T·∫°o c·ªôt t·ªïng l∆∞·ª£t t∆∞∆°ng t√°c
df_post['interaction'] = df_post['reaction_count'] + df_post['comments'] + df_post['shares']
df_post
```

```python
# B√†i post c√≥ t·ªïng l∆∞·ª£t t∆∞∆°ng t√°c cao nh·∫•t
highest_interaction_post = df_post.loc[df_post['interaction'] == df_post['interaction'].max()]
highest_interaction_post
```

## Th·ªùi gian m√† page th∆∞·ªùng xuy√™n ƒëƒÉng b√†i trong ng√†y

```python
list_hour = df_post['hour_time'].value_counts().sort_index()
list_hour
```

```python
# S·ªë b√†i post ƒëƒÉng theo th·ªùi ƒëi·ªÉm trong ng√†y
ax = list_hour.plot(kind='bar', figsize=(16,5))
ax.set_xlabel('Gi√≤ ƒëƒÉng b√†i')
ax.set_ylabel('S·ªë b√†i ƒë√£ ƒëƒÉng')
```

```python
# L∆∞·ª£t t∆∞∆°ng t√°c trong c√°c khung gi·ªù
reaction_hour = df_post.groupby('hour_time').sum('interaction')
reaction_hour
```

```python
ax = reaction_hour[['reaction_count', 'comments', 'shares']].plot.bar(stacked=True, figsize=(16,8))
ax.set_xlabel('Gi·ªù ƒëƒÉng b√†i')
ax.set_ylabel('T∆∞∆°ng t√°c')
```

## L∆∞·ª£t t∆∞∆°ng t√°c v·ªõi nh·ªØng post c√≥ video v√† h√¨nh ·∫£nh

```python
# L·∫•y nh·ªØng post c√≥ ·∫£nh
list_image = df_post.loc[df_post['image'] != 0]
```

```python
# L·∫•y nh·ªØng post c√≥ video
list_video = df_post.loc[df_post['video'] != 0]
```

```python
# L∆∞·ª£ng t∆∞∆°ng t√°c trung b√¨nh c·ªßa c√°c post
mean_interaction = df_post['interaction'].mean()
mean_interaction
```

```python
# L∆∞·ª£ng t∆∞∆°ng t√°c v·ªõi nh·ªØng post c√≥ video
x = list_video[['th√≠ch', 'y√™u th√≠ch', 'th∆∞∆°ng th∆∞∆°ng', 'haha', 'wow', 'bu·ªìn', 'ph·∫´n n·ªô', 'comments', 'shares']]
ax = x.plot(kind='bar', figsize=(20,8), stacked=True)
ax.axhline(mean_interaction, color='red')
ax.set_xlabel('Video post')
ax.set_ylabel('Interaction')
```

```python
# L∆∞·ª£t t∆∞∆°ng t√°c v·ªõi nh·ªØng post c√≥ ·∫£nh
x = list_image[['th√≠ch', 'y√™u th√≠ch', 'th∆∞∆°ng th∆∞∆°ng', 'haha', 'wow', 'bu·ªìn', 'ph·∫´n n·ªô', 'comments', 'shares']]
ax = x.plot(kind='bar', figsize=(20,8), stacked=True)
ax.axhline(mean_interaction, color='red')
ax.set_xlabel('Image post')
ax.set_ylabel('Interaction')
```

```python
# L∆∞·ª£t t∆∞∆°ng t√°c v·ªõi nh·ªØng post ch·ªâ c√≥ text
x = list_only_text[['th√≠ch', 'y√™u th√≠ch', 'th∆∞∆°ng th∆∞∆°ng', 'haha', 'wow', 'bu·ªìn', 'ph·∫´n n·ªô', 'comments', 'shares']]
ax = x.plot(kind='bar', figsize=(20,8), stacked=True)
ax.axhline(mean_interaction, color='red')
ax.set_xlabel('Only_text post')
ax.set_ylabel('Interaction')
```

```python
# L·∫•y t·∫•t c·∫£ c√°c b√†i vi·∫øt
text = ""
for i in df_post['post_text']:
  text=text+i+"\n"
```

```python
# Import WordCloud library
from wordcloud import WordCloud
from pathlib import Path
import os
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
```

```python
# Nh·ªØng t·ª´ xu·∫•t hi·ªán nhi·ªÅu nh·∫•t
texto = text
wordcloud = WordCloud(stopwords=STOPWORDS,
                          background_color='white',
                      max_words=300,
                      width=500, height=300
                         ).generate(texto)
plt.figure(figsize=(20,10))
plt.clf()
plt.imshow(wordcloud)
plt.axis('off')
plt.show()
```
